```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Generate synthetic data
np.random.seed(0)  # For reproducibility

# Generate heights
heights = np.random.normal(loc=170, scale=10, size=100)

# Generate weights based on heights with some noise
weights = 0.6 * heights - 25 + np.random.normal(loc=0, scale=5, size=100)
```

```python

heights_normalized = (heights - np.mean(heights)) / np.std(heights)
weights_normalized = (weights - np.mean(weights)) / np.std(weights) 

# Plot the data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(heights, weights, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(heights_normalized, weights_normalized, color='orange')
plt.title('Normalized Data')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
data = np.column_stack((heights, weights))
pca = PCA(n_components=2)
pca.fit(data)
transformed_data = pca.transform(data)

heights_pca, weights_pca = zip(*transformed_data)

heights_normalized = (heights_pca - np.mean(heights_pca)) / np.std(heights_pca)
weights_normalized = (weights_pca - np.mean(weights_pca)) / np.std(weights_pca) 

# Plot the data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(heights, weights, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(heights_normalized, weights_normalized, color='orange')
plt.title('Normalized Data')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```
