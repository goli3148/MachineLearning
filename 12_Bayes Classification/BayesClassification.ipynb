The predicted class is the one with the highest posterior probability.</p></li></ol>

```python
class BayseClassification:
    xdata: NDArray
    ydata: NDArray
    def __init__(self, xdata, ydata) -> None:
        self.xdata = xdata
        self.ydata = ydata
        
    def fit(self):
        self.classes = np.unique(self.ydata)
        self.classes_parametes = {}
        for c in self.classes:
            X_c = self.xdata[self.ydata == c]
            mean = np.mean(X_c, axis=0)
            var  = np.var(X_c, axis=0)
            self.classes_parametes[c] = {'m':mean, 's':var}
            
    def posterior(self, data):
        probabilites = {}
        for c in self.classes:
            class_params = self.classes_parametes[c]
            likelihood_class = np.sum(self.likelihood(class_params['m'], class_params['s'], data))
            prior_class = self.prior(c)
            probabilites[c] = likelihood_class + prior_class
        return probabilites
            
    def likelihood(self, mean, std, data):
        exponent = np.exp(-(np.power(data - mean, 2) / (2 * np.power(std, 2))))
        return np.log((1 / (np.sqrt(2 * np.pi) * std)) * exponent)
    
    def prior(self, label):
        return np.log((self.ydata == label).sum() / len(self.ydata))
        
    def prediction(self, Data):
        predictions = []
        for data in Data:
            pro:Dict = self.posterior(data)
            predicted = max(pro, key=pro.get)
            predictions.append(predicted)
        return predictions
    
```

```python
iris = load_iris()
X = iris.data
y = iris.target
# simplify to a two-class
# class 0 and class 1 will be first class and class 2 will be second class
y = np.array([0 if label==0 or label==1 else 1 for label in y])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
BC = BayseClassification(X_train, y_train)
BC.fit()
yPredict = BC.prediction(X_test)
res = 0
for index in range(len(yPredict)):
    res = res+1 if yPredict[index] == y_test[index] else res
print(res/len(yPredict))
```
