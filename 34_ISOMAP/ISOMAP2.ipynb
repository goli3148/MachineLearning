```python
from sklearn.datasets import make_s_curve
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt
```

```python
n_neighbors = 12
n_components = 2
n_samples = 1500
S_points, S_color = make_s_curve(n_samples, random_state=0)
isomap = Isomap(n_neighbors=n_neighbors, n_components=n_components, p=1)
S_isomap = isomap.fit_transform(S_points)
plt.figure(figsize=(8, 6))
scatter = plt.scatter(S_isomap[:, 0], S_isomap[:, 1], c=S_color, cmap='viridis', s=15, alpha=0.7)
plt.colorbar(scatter, label="S-curve Color Gradient")
plt.xlabel("MDS Component 1")
plt.ylabel("MDS Component 2")
plt.title("MDS Projection of S-curve Dataset")
plt.show()
```
